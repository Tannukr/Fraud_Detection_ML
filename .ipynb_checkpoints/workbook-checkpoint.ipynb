````python
df = df.drop(columns=["nameOrig","nameDest"])
````

````python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
````

````python
X = df.drop(columns=['isFraud'])
y = df["isFraud"]
````

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
````

````python
categorical = ["type"]
numeric = [col for col in X.columns if col not in categorical]
````

````python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numeric),
        ("cat", OneHotEncoder(drop="first", handle_unknown="ignore"), categorical)
    ],
    remainder="drop"   # ensures NO extra columns
)
````

````python
X_train_transformed = preprocessor.fit_transform(X_train)
X_test_transformed  = preprocessor.transform(X_test)
````

````python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42, sampling_strategy=0.1)  
# sampling_strategy=0.1 → fraud cases will be 10% of normal

X_train_resampled, y_train_resampled = smote.fit_resample(X_train_transformed, y_train)

print("Before SMOTE:", y_train.value_counts())
print("After SMOTE:", y_train_resampled.value_counts())
````

````python
model = LogisticRegression(max_iter=1000, class_weight="balanced")
model.fit(X_train_resampled, y_train_resampled)
````

````python
y_pred = model.predict(X_test_transformed)
y_proba = model.predict_proba(X_test_transformed)[:, 1]

print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("ROC-AUC Score:", roc_auc_score(y_test, y_proba))
````

````python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, roc_auc_score

# Support Vector Machine (SVM)
svm = SVC(
    kernel="rbf",              # "linear" or "rbf"
    C=1.0,                     # regularization strength
    class_weight="balanced",
    probability=True,           # needed for ROC-AUC
    max_iter=1000,              # upper limit for iterations
    tol=1e-3,
    random_state=42
)

svm.fit(X_train_resampled, y_train_resampled)

y_pred_svm = svm.predict(X_test_transformed)

print("Support Vector Machine (SVM) with Regularization")
print(classification_report(y_test, y_pred_svm))
print("ROC-AUC:", roc_auc_score(y_test, svm.predict_proba(X_test_transformed)[:, 1]))

````
